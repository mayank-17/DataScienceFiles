```python
import seaborn as sns
```

```python
myTitanic = sns.load_dataset('titanic')
```

```python
myTitanic.head()
```

```python
sns.distplot(myTitanic['fare']) #Distribution Plotb
```

```python
sns.distplot(myTitanic['fare'], kde = False, bins = 10)
```

```python
sns.kdeplot(myTitanic['fare']) # Kernel Density Estimation
```

```python
sns.jointplot(x = 'age', y = 'fare', data = myTitanic)
```

```python
sns.jointplot(x = 'age', y = 'fare', data = myTitanic, kind = 'hex')
```

```python
sns.jointplot(x = 'age', y = 'fare', data = myTitanic, kind = 'reg')
```

```python
sns.jointplot(x = 'age', y = 'fare', data = myTitanic, kind = 'scatter')
```
